```python
#import pandas library
import pandas as pd
import numpy as np
```

```python
#import food data
fData = pd.read_csv('Food_Products - Food_Products.csv')
```

```python
#since the table is large, I created a menu to pick what is desired
inMenu = True
while inMenu:
    print("""
    (a) Find products with high saturated fats 
    (b) Find products with high added sugars
    (c) Find products with more green vegetables
    (x) exit
    """)
    answer = input("Please Select an Option: ")
    if answer == 'a':
        print ("(a) the following products have high saturated fats")
        highsatfat=fData[fData['Saturated_Fats']> 13][['Display_Name','Saturated_Fats']].sort_values(by=['Saturated_Fats','Display_Name'],ascending=[False,True])
        print(highsatfat)
    elif answer == 'b':
        highaddedsugars=fData[fData['Added_Sugars']> 76.7][['Display_Name','Added_Sugars']].sort_values(by=['Added_Sugars','Display_Name'],ascending=[False,True])
        print(highaddedsugars)
    elif answer == 'c':
        greenveg=fData[fData['Drkgreen_Vegetables']> 0.1][['Display_Name','Drkgreen_Vegetables']].sort_values(by=['Drkgreen_Vegetables','Display_Name'],ascending=[False,True])
        print(greenveg)
    elif answer == 'x':
        break
    else:
        print ("unknown input, please try again")
   
```
